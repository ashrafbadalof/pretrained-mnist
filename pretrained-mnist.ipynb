```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import tensorflow as tf
import keras
from keras import layers
from keras.layers import Dense
```

```python
data = tf.keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = data.load_data()
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
np.unique(y_train)
```

```python
num_rows = 5
num_cols = 5
fig = plt.figure(figsize=(10,8))
for image_idx, image in enumerate(X_train[:num_rows*num_cols]):
    ax = fig.add_subplot(num_rows, num_cols, image_idx +1)
    ax.imshow(image, cmap = 'gray')
    ax.set_title(y_train[image_idx])
    ax.axis("off")
    plt.tight_layout()
```

```python
X_train_sc = X_train/255.0
X_test_sc = X_test/255.0
```

```python
selected_digits = [0, 1, 2, 3, 4]
selected_indices = np.isin(y_train, selected_digits)
X_train_selected = X_train_sc[selected_indices]
y_train_selected = y_train[selected_indices]
```

```python
np.unique(y_train_selected)
```

```python
X_train_selected.shape
```

```python
num_row = 2
num_col = 5
fig = plt.figure(figsize=(5,4))
for image_idx, image in enumerate(X_train_selected[:num_row*num_col]):
    ax = fig.add_subplot(num_row, num_col, image_idx + 1)
    ax.imshow(image, cmap = "gray")
    ax.set_title(y_train_selected[image_idx])
    ax.axis("off")
    plt.tight_layout()
# showing only the numbers in the range of 0-4
```

```python
X_train_last, X_valid = X_train_selected[:25000], X_train_selected[25000:]
y_train_last, y_valid = y_train_selected[:25000], y_train_selected[25000:]
```

```python
print(X_train_last.shape)
print(X_valid.shape)
y_valid.shape
```

```python
early_stopping = keras.callbacks.EarlyStopping(patience = 10)
model_checkpoint = keras.callbacks.ModelCheckpoint("my_mnist1.h5", save_best_only = True)
callbacks = [early_stopping, model_checkpoint]
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = [28,28]),
    keras.layers.Dense(100, kernel_initializer = "he_normal", activation="elu"),
    keras.layers.Dense(100, kernel_initializer = "he_normal", activation="elu"),
    keras.layers.Dense(100, kernel_initializer = "he_normal", activation="elu"),
    keras.layers.Dense(100, kernel_initializer = "he_normal", activation="elu"),
    keras.layers.Dense(100, kernel_initializer = "he_normal", activation="elu"),
    keras.layers.Dense(5, activation="softmax")
])
```

```python
model.compile(
    optimizer = "adam",
    loss = "sparse_categorical_crossentropy",
    metrics = ["accuracy"]
)
```

```python
history = model.fit(X_train_last, y_train_last, validation_data = (X_valid, y_valid), batch_size = 32, epochs = 50, callbacks = callbacks)
```

```python
p_model = keras.models.load_model("my_mnist1.h5")
p_model.evaluate(X_valid, y_valid)
```
